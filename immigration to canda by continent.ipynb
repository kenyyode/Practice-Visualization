```python
import pandas as pd 
import numpy as np 
import matplotlib as mpl 
import matplotlib.pyplot as plt 

%matplotlib inline 
```

```python
df = pd.read_csv("Cleaned_data.csv")

df.head(3)
```

```python
df_canada = df
```

```python
df_canada.columns
```

```python
df_canada.rename(columns={"Unnamed: 0": "Country"}, inplace=True)
```

```python
df_canada.head(3)
```

```python
df_= df_canada.groupby('Continent').sum()
```

```python
df_
```

```python
years = list(map(str, range(1980, 2014)))

df_asia = df_.loc["Asia", years ]
```

```python
#df_asia.shape

df_asia
```

```python
df_asia.plot(kind='bar')

plt.title("Immigrants from Iceland to canada from 1980 to 2014")
plt.xlabel("years")
plt.ylabel("Number of immigrants")

plt.show()
```

```python
myex = [0.04] * len(df_)
df_["Total"].plot(kind="pie", explode=myex)
plt.title("Immigration to canada by continent")
plt.show()
```
