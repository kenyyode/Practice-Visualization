```python
import pandas as pd 
import numpy as np 
import matplotlib as mpl 
import matplotlib.pyplot as plt 

%matplotlib inline 
```

```python
df = pd.read_csv("Cleaned_data.csv")

df.head(3)
```

```python
df_canada = df
```

```python
## this sorts values based on the country with the most immigrations to canada. 
df_canada.sort_values(['Total'], axis=0, ascending=False, inplace=True)
```

```python
df_canada.head(4)
```

```python
df_canada.columns
```

```python
df_canada.rename(columns={"Unnamed: 0": "Country"}, inplace=True)
```

```python
df_canada.head(3)
```

```python
df_canada.set_index('Country', inplace=True)
```

```python
df_canada.index.unique()
```

```python
years = list(map(str, range(1980, 2014)))

df_iceland = df_canada.loc["Iceland", years ]
```

```python
df_iceland.shape

df_iceland.loc["1991"]
```

```python
df_iceland.plot(kind='bar')

plt.title("Immigrants from Iceland to canada from 1980 to 2014")
plt.xlabel("years")
plt.ylabel("Number of immigrants")

plt.show()
```
